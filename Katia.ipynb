```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generar datos para la distribucion normal

media = 70 # Media
desv_std =4 # Desviacion estandar
num_datos = 500

#generamos los datos de manera aleatoria y los ponemos en una lista (array)
datos = np.random.normal(media, desv_std, num_datos)
datos[0:50] # Mostrar los primeros 50
```

```python
# Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, color='b') #Histograma de los datos
plt.title('Distribucion Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
# # Crear un rango de valores para crear una densidad
x = np.linspace(media -  3 * desv_std, media + 3 *desv_std, num_datos)

# Calcular la PDF de la distribucion normal
pdf = 1 / (desv_std * np.sqrt(2 * np.pi)) * np.exp(-(x - media)**2 / (2 * desv_std **2))
```

```python
pdf.size, num_datos
```

```python
# Para no imprimir toda la lista de valores (500), desplegamos los primeros 50
print("Imprimir los primeros 50 Valores de x", x[0:50], " , y los imprimeros 50 pdf", pdf[0:50])
```

```python
# Visualizar la PDF
plt.figure(figsize=(6,4))
plt.plot(x, pdf, label='Normal:(media:{}; ds:{})'.format(np.round(media, 2), np.round(desv_std,2 )))
plt.title('Funcion de Densidad de Probabilidad (PDF) - Distribucion Normal')
plt.xlabel('Pesos en Kg. de los estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Graficar histograma
plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')
plt.plot( x, pdf, 'k', linewidth=2, )

plt.title('Distribucion Normal ')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 10
p = 0.10

valores_x= np.arange(0, n+1)
probabilidades_x = [binom.pmf(x, n, p) for x in valores_x]

print("Valores de x\t probabilidades")
for x, probabilidad in zip(valores_x, probabilidades_x):
  print(f"{x}\t\t{probabilidad}")
```

```python
# Graficar el histograma de la distribución binomial
plt.figure(figsize=(8, 6))
plt.bar(valores_x, probabilidades_x, color='skyblue')
plt.title('Distribución Binomial')
plt.xlabel('Valor de X')
plt.ylabel('Probabilidad')
plt.xticks(valores_x)
plt.grid(True)
plt.show()
```

#Poison#

```python
from scipy.stats import poisson

# Parámetros de la distribución de Poisson
media_accidentes = 2

# Parte 1: Probabilidad de que no haya accidentes en 1 semana
probabilidad_0_accidentes = poisson.pmf(0, media_accidentes)

print("1. Probabilidad de que no haya accidentes en 1 semana:", probabilidad_0_accidentes)

# Parte 2: Probabilidad de que a lo sumo haya tres accidentes en 2 semanas
# Convertimos la media a una media para 2 semanas
media_2_semanas = media_accidentes * 2

# Calculamos la probabilidad acumulada hasta 3 accidentes en 2 semanas
probabilidad_3_o_menos_accidentes = poisson.cdf(3, media_2_semanas)

print("2. Probabilidad de que a lo sumo haya tres accidentes en 2 semanas:", probabilidad_3_o_menos_accidentes)
```
