# 1. Carga y Exploración de Datos (Grupo 1)#

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
grupomewtwo = pd.read_csv("/content/grupo2.csv")

# Exploración inicial
print(grupomewtwo.info())
print(grupomewtwo.describe())

# Histograma de calificaciones
plt.figure(figsize=(6, 4))
sns.histplot(grupomewtwo["Calificacion"], kde=True, bins=10, color='skyblue')
plt.title("Histograma de Calificaciones")
plt.xlabel("Calificación")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

# 2. Modelo de Regresión Lineal #

```python
from sklearn.linear_model import LinearRegression

# Variables
X = grupomewtwo[["HorasEstudio"]]
y = grupomewtwo["Calificacion"]

# Modelo
lm = LinearRegression()
lm.fit(X, y)

# Parámetros
pendiente = lm.coef_[0]
interseccion = lm.intercept_
ecuacion = f"y = {pendiente:.2f}x + {interseccion:.2f}"

print(f"Ecuación de la línea de regresión: {ecuacion}")
```

# 3. Cálculo de la Correlación de Pearson #

```python
correlacion = grupomewtwo[["HorasEstudio", "Calificacion"]].corr().iloc[0,1]
print(f"Coeficiente de correlación de Pearson: {correlacion:.4f}")
```

# 4. Visualización de Resultados #

```python
plt.figure(figsize=(6, 4))
sns.regplot(x="HorasEstudio", y="Calificacion", data=grupomewtwo, scatter=True, color="blue")
plt.title("Diagrama de dispersión con línea de regresión")
plt.xlabel("Horas de estudio")
plt.ylabel("Calificación obtenida")
plt.grid(True)
plt.show()
```

# 5. Interpretación de Resultados #

¿Interpretación del coeficiente de regresión?

La ecuación de la línea de regresión es:

Calificacion = 4.22 x Horas de estudio + 15.31

Esto significa que por cada hora adicional de estudio, la calificación esperada aumenta en promedio 4.22 puntos. El modelo sugiere una relación positiva entre el tiempo dedicado al estudio y el rendimiento académico.

2. ¿Interpretación del coeficiente de correlación?

El coeficiente de correlación de Pearson es 0.8620, lo que indica una correlación positiva fuerte entre las horas de estudio y las calificaciones. Esto quiere decir que, en general, a mayor cantidad de horas de estudio, mayores son las calificaciones obtenidas.

3. ¿Es estadísticamente significativo el modelo?

Sí, debido a que el coeficiente de correlación es alto y positivo (0.8620), el modelo sugiere una relación lineal clara entre ambas variables. Además, la pendiente es considerable (> 0), lo cual refuerza que la variable horas de estudio es un buen predictor del rendimiento académico.

4. ¿Qué tan confiable es el modelo para predecir calificaciones?

Es un modelo bastante confiable. Con una pendiente significativa y una fuerte correlación, el modelo es útil para predecir calificaciones en función del tiempo de estudio. Aunque siempre habrá variabilidad individual, el patrón general es claro y consistente.
