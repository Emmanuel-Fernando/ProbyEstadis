# 1. Carga y Exploraci√≥n de Datos (Grupo 1)#

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
grupomewtwo = pd.read_csv("/content/grupo2.csv")

# Exploraci√≥n inicial
print(grupomewtwo.info())
print(grupomewtwo.describe())

# Histograma de calificaciones
plt.figure(figsize=(6, 4))
sns.histplot(grupomewtwo["Calificacion"], kde=True, bins=10, color='skyblue')
plt.title("Histograma de Calificaciones")
plt.xlabel("Calificaci√≥n")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

# 2. Modelo de Regresi√≥n Lineal #

```python
from sklearn.linear_model import LinearRegression

# Variables
X = grupomewtwo[["HorasEstudio"]]
y = grupomewtwo["Calificacion"]

# Modelo
lm = LinearRegression()
lm.fit(X, y)

# Par√°metros
pendiente = lm.coef_[0]
interseccion = lm.intercept_
ecuacion = f"y = {pendiente:.2f}x + {interseccion:.2f}"

print(f"Ecuaci√≥n de la l√≠nea de regresi√≥n: {ecuacion}")
```

# 3. C√°lculo de la Correlaci√≥n de Pearson #

```python
correlacion = grupomewtwo[["HorasEstudio", "Calificacion"]].corr().iloc[0,1]
print(f"Coeficiente de correlaci√≥n de Pearson: {correlacion:.4f}")
```

# 4. Visualizaci√≥n de Resultados #

```python
plt.figure(figsize=(6, 4))
sns.regplot(x="HorasEstudio", y="Calificacion", data=grupomewtwo, scatter=True, color="blue")
plt.title("Diagrama de dispersi√≥n con l√≠nea de regresi√≥n")
plt.xlabel("Horas de estudio")
plt.ylabel("Calificaci√≥n obtenida")
plt.grid(True)
plt.show()
```

# 5. Interpretaci√≥n de Resultados #

¬øInterpretaci√≥n del coeficiente de regresi√≥n?

La ecuaci√≥n de la l√≠nea de regresi√≥n es:

Calificacion = 4.22 x Horas de estudio + 15.31

Esto significa que por cada hora adicional de estudio, la calificaci√≥n esperada aumenta en promedio 4.22 puntos. El modelo sugiere una relaci√≥n positiva entre el tiempo dedicado al estudio y el rendimiento acad√©mico.

2. ¬øInterpretaci√≥n del coeficiente de correlaci√≥n?

El coeficiente de correlaci√≥n de Pearson es 0.8620, lo que indica una correlaci√≥n positiva fuerte entre las horas de estudio y las calificaciones. Esto quiere decir que, en general, a mayor cantidad de horas de estudio, mayores son las calificaciones obtenidas.

3. ¬øEs estad√≠sticamente significativo el modelo?

S√≠, debido a que el coeficiente de correlaci√≥n es alto y positivo (0.8620), el modelo sugiere una relaci√≥n lineal clara entre ambas variables. Adem√°s, la pendiente es considerable (> 0), lo cual refuerza que la variable horas de estudio es un buen predictor del rendimiento acad√©mico.

4. ¬øQu√© tan confiable es el modelo para predecir calificaciones?

Es un modelo bastante confiable. Con una pendiente significativa y una fuerte correlaci√≥n, el modelo es √∫til para predecir calificaciones en funci√≥n del tiempo de estudio. Aunque siempre habr√° variabilidad individual, el patr√≥n general es claro y consistente.
