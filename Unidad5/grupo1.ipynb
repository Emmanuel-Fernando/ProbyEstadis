# 1. Carga y Exploraci√≥n de Datos (Grupo 1)#

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
grupoone = pd.read_csv("/content/grupo1.csv")

# Exploraci√≥n inicial
print(grupoone.info())
print(grupoone.describe())

# Histograma de calificaciones
plt.figure(figsize=(6, 4))
sns.histplot(grupoone["Calificacion"], kde=True, bins=10, color='skyblue')
plt.title("Histograma de Calificaciones")
plt.xlabel("Calificaci√≥n")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

# 2. Modelo de Regresi√≥n Lineal #

```python
from sklearn.linear_model import LinearRegression

# Variables
X = grupoone[["HorasEstudio"]]
y = grupoone["Calificacion"]

# Modelo
lm = LinearRegression()
lm.fit(X, y)

# Par√°metros
pendiente = lm.coef_[0]
interseccion = lm.intercept_
ecuacion = f"y = {pendiente:.2f}x + {interseccion:.2f}"

print(f"Ecuaci√≥n de la l√≠nea de regresi√≥n: {ecuacion}")
```

# 3. C√°lculo de la Correlaci√≥n de Pearson #

```python
correlacion = grupoone[["HorasEstudio", "Calificacion"]].corr().iloc[0,1]
print(f"Coeficiente de correlaci√≥n de Pearson: {correlacion:.4f}")
```

# 4. Visualizaci√≥n de Resultados #

```python
plt.figure(figsize=(6, 4))
sns.regplot(x="HorasEstudio", y="Calificacion", data=grupoone, scatter=True, color="blue")
plt.title("Diagrama de dispersi√≥n con l√≠nea de regresi√≥n")
plt.xlabel("Horas de estudio")
plt.ylabel("Calificaci√≥n obtenida")
plt.grid(True)
plt.show()
```

# 5. Interpretaci√≥n de Resultados #

¬øInterpretaci√≥n del coeficiente de regresi√≥n?

La pendiente indica cu√°nto se espera que cambie la calificaci√≥n por cada hora adicional de estudio. En este caso, como la pendiente es cercana a 0 o negativa, sugiere que no hay un impacto positivo fuerte del estudio sobre la calificaci√≥n.

2. ¬øInterpretaci√≥n del coeficiente de correlaci√≥n?

El coeficiente de correlaci√≥n de Pearson indica la fuerza y direcci√≥n de la relaci√≥n. Si el valor es cercano a 0 (como en este caso, ~-0.05), hay muy poca o ninguna relaci√≥n lineal entre las horas de estudio y la calificaci√≥n.

3. ¬øEs estad√≠sticamente significativo el modelo?

Dado que la correlaci√≥n es baja y la pendiente no es considerable, el modelo no es estad√≠sticamente significativo para predecir calificaciones con base en el tiempo de estudio.

4. ¬øQu√© tan confiable es el modelo para predecir calificaciones?

El modelo no es confiable, ya que los resultados muestran una relaci√≥n d√©bil entre las variables.
