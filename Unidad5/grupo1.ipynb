# 1. Carga y Exploración de Datos (Grupo 1)#

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
grupoone = pd.read_csv("/content/grupo1.csv")

# Exploración inicial
print(grupoone.info())
print(grupoone.describe())

# Histograma de calificaciones
plt.figure(figsize=(6, 4))
sns.histplot(grupoone["Calificacion"], kde=True, bins=10, color='skyblue')
plt.title("Histograma de Calificaciones")
plt.xlabel("Calificación")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

# 2. Modelo de Regresión Lineal #

```python
from sklearn.linear_model import LinearRegression

# Variables
X = grupoone[["HorasEstudio"]]
y = grupoone["Calificacion"]

# Modelo
lm = LinearRegression()
lm.fit(X, y)

# Parámetros
pendiente = lm.coef_[0]
interseccion = lm.intercept_
ecuacion = f"y = {pendiente:.2f}x + {interseccion:.2f}"

print(f"Ecuación de la línea de regresión: {ecuacion}")
```

# 3. Cálculo de la Correlación de Pearson #

```python
correlacion = grupoone[["HorasEstudio", "Calificacion"]].corr().iloc[0,1]
print(f"Coeficiente de correlación de Pearson: {correlacion:.4f}")
```

# 4. Visualización de Resultados #

```python
plt.figure(figsize=(6, 4))
sns.regplot(x="HorasEstudio", y="Calificacion", data=grupoone, scatter=True, color="blue")
plt.title("Diagrama de dispersión con línea de regresión")
plt.xlabel("Horas de estudio")
plt.ylabel("Calificación obtenida")
plt.grid(True)
plt.show()
```

# 5. Interpretación de Resultados #

¿Interpretación del coeficiente de regresión?

La pendiente indica cuánto se espera que cambie la calificación por cada hora adicional de estudio. En este caso, como la pendiente es cercana a 0 o negativa, sugiere que no hay un impacto positivo fuerte del estudio sobre la calificación.

2. ¿Interpretación del coeficiente de correlación?

El coeficiente de correlación de Pearson indica la fuerza y dirección de la relación. Si el valor es cercano a 0 (como en este caso, ~-0.05), hay muy poca o ninguna relación lineal entre las horas de estudio y la calificación.

3. ¿Es estadísticamente significativo el modelo?

Dado que la correlación es baja y la pendiente no es considerable, el modelo no es estadísticamente significativo para predecir calificaciones con base en el tiempo de estudio.

4. ¿Qué tan confiable es el modelo para predecir calificaciones?

El modelo no es confiable, ya que los resultados muestran una relación débil entre las variables.
