# 1. Carga y Exploración de Datos (Grupo 3) #

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
grupotripas = pd.read_csv("/content/grupo3.csv")

# Exploración inicial
print(grupotripas.info())
print(grupotripas.describe())

# Histograma de calificaciones
plt.figure(figsize=(6, 4))
sns.histplot(grupotripas["Calificacion"], kde=True, bins=10, color='skyblue')
plt.title("Histograma de Calificaciones")
plt.xlabel("Calificación")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

# 2. Modelo de Regresión Lineal #

```python
from sklearn.linear_model import LinearRegression

# Variables
X = grupotripas[["HorasEstudio"]]
y = grupotripas["Calificacion"]

# Modelo
lm = LinearRegression()
lm.fit(X, y)

# Parámetros
pendiente = lm.coef_[0]
interseccion = lm.intercept_
ecuacion = f"y = {pendiente:.2f}x + {interseccion:.2f}"

print(f"Ecuación de la línea de regresión: {ecuacion}")
```

# 3. Cálculo de la Correlación de Pearson #

```python
correlacion = grupotripas[["HorasEstudio", "Calificacion"]].corr().iloc[0,1]
print(f"Coeficiente de correlación de Pearson: {correlacion:.4f}")
```

# 4. Visualización de Resultados #

```python
plt.figure(figsize=(6, 4))
sns.regplot(x="HorasEstudio", y="Calificacion", data=grupotripas, scatter=True, color="blue")
plt.title("Diagrama de dispersión con línea de regresión")
plt.xlabel("Horas de estudio")
plt.ylabel("Calificación obtenida")
plt.grid(True)
plt.show()
```

# 5. Interpretación de Resultados #

¿Interpretación del coeficiente de regresión?

La ecuación de la línea de regresión es:

Calificacion=5.29 x Horas de estudio + 0.11

Esto significa que por cada hora adicional de estudio, la calificación esperada aumenta en promedio 5.29 puntos. Este valor muestra una relación muy positiva entre el tiempo de estudio y las calificaciones: más estudio se traduce en un mejor desempeño académico.

2. ¿Interpretación del coeficiente de correlación?

El coeficiente de correlación de Pearson es 0.9422, lo que indica una correlación positiva muy fuerte. Esto sugiere que la relación entre horas de estudio y calificaciones es casi perfectamente lineal: los estudiantes que estudian más obtienen consistentemente mejores resultados.

3. ¿Es estadísticamente significativo el modelo?

Sí, el modelo es altamente significativo. Un coeficiente de correlación de 0.9422 implica una relación lineal muy clara. Además, la pendiente de 5.29 muestra un impacto fuerte del estudio sobre la calificación. Esto valida la utilidad del modelo para representar la relación entre las variables.

4. ¿Qué tan confiable es el modelo para predecir calificaciones?

Este modelo es muy confiable para hacer predicciones. Gracias a su alta correlación y una pendiente bien definida, el modelo puede estimar las calificaciones con buen grado de precisión, siempre considerando que pueden existir otros factores no incluidos en el análisis.
