# 1. Carga y Exploraci√≥n de Datos (Grupo 3) #

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
grupotripas = pd.read_csv("/content/grupo3.csv")

# Exploraci√≥n inicial
print(grupotripas.info())
print(grupotripas.describe())

# Histograma de calificaciones
plt.figure(figsize=(6, 4))
sns.histplot(grupotripas["Calificacion"], kde=True, bins=10, color='skyblue')
plt.title("Histograma de Calificaciones")
plt.xlabel("Calificaci√≥n")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

# 2. Modelo de Regresi√≥n Lineal #

```python
from sklearn.linear_model import LinearRegression

# Variables
X = grupotripas[["HorasEstudio"]]
y = grupotripas["Calificacion"]

# Modelo
lm = LinearRegression()
lm.fit(X, y)

# Par√°metros
pendiente = lm.coef_[0]
interseccion = lm.intercept_
ecuacion = f"y = {pendiente:.2f}x + {interseccion:.2f}"

print(f"Ecuaci√≥n de la l√≠nea de regresi√≥n: {ecuacion}")
```

# 3. C√°lculo de la Correlaci√≥n de Pearson #

```python
correlacion = grupotripas[["HorasEstudio", "Calificacion"]].corr().iloc[0,1]
print(f"Coeficiente de correlaci√≥n de Pearson: {correlacion:.4f}")
```

# 4. Visualizaci√≥n de Resultados #

```python
plt.figure(figsize=(6, 4))
sns.regplot(x="HorasEstudio", y="Calificacion", data=grupotripas, scatter=True, color="blue")
plt.title("Diagrama de dispersi√≥n con l√≠nea de regresi√≥n")
plt.xlabel("Horas de estudio")
plt.ylabel("Calificaci√≥n obtenida")
plt.grid(True)
plt.show()
```

# 5. Interpretaci√≥n de Resultados #

¬øInterpretaci√≥n del coeficiente de regresi√≥n?

La ecuaci√≥n de la l√≠nea de regresi√≥n es:

Calificacion=5.29 x Horas de estudio + 0.11

Esto significa que por cada hora adicional de estudio, la calificaci√≥n esperada aumenta en promedio 5.29 puntos. Este valor muestra una relaci√≥n muy positiva entre el tiempo de estudio y las calificaciones: m√°s estudio se traduce en un mejor desempe√±o acad√©mico.

2. ¬øInterpretaci√≥n del coeficiente de correlaci√≥n?

S√≠, el modelo es altamente significativo. Un coeficiente de correlaci√≥n de 0.9422 implica una relaci√≥n lineal muy clara.
