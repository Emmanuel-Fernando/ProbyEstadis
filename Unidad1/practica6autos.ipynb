```python
import math
import matplotlib.pyplot as plt
```

```python
#Calcular el numero de intervalos usando la regla de struges
n = len(autos.price)
k = int(1 + math.log2(n))
print("Valor de n: ", n, " \nNumero de intervalos: ", k)
```

```python
plt.hist(autos.price, bins=k)
plt.title ("Histograma de precios basado en la funion de matplotlib con valores estandar")
```

```python
plt.hist(autos['price'], bins= k, alpha=0.5)
plt.title("Histograma de precios basado en la regla de Sturgles")
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
```
