```python
media = np.mean(autos.price)
print(media)
```

```python
mediana = median(autos.price)
print(mediana)
```

```python
autos.price.describe()
```

```python
import math
import matplotlib.pyplot as plt
```

```python
#Calcular el numero de intervalos usando la regla de struges
n = len(autos.price)
k = int(1 + math.log2(n))
print("Valor de n: ", n, " \nNumero de intervalos: ", k)
```

```python
plt.hist(autos.price, bins=k)
plt.title ("Histograma de precios basado en la funion de matplotlib con valores estandar")
```

```python
plt.hist(autos['price'], bins= k, alpha=0.5)
plt.title("Histograma de precios basado en la regla de Sturgles")
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
```

```python
plt.figure(figsize= (8,6) )
plt.hist(autos['price'],bins=10,alpha=0.2, color="b", edgecolor="black")
plt.axvline(media,color="b",linestyle="dashed",linewidth=2,label=f"Media: {media:.2f}")
plt.axvline(mediana,color="r",linestyle="dashed",linewidth=2,label=f"Mediana: {mediana:.2f}")
plt.axvline(moda1,color="y",linestyle="dashed",linewidth=2,label=f"Moda: {moda1:.2f}")

plt.axvline(media - desviacion_estandar,color= "black",linestyle= "dashed",linewidth= 2,label= f"Desv.Estandar: {desviacion_estandar:.2f}")
plt.axvline(media+desviacion_estandar,color="black",linestyle="dashed",linewidth=2)

plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.title("Produccion diaria de la fabrica 1")
plt.legend()
plt.show()
```

#CONCLUSIONES
### La mayor cantidad de autos se encuentran con un precio entre 5 000 y 10 000 mientras que hay menor cantidad de autos de precio 25 000 - 30 000 y +40 000

```python
plt.boxplot(autos['price'])
plt.title('Precio')
plt.show()
```

```python
tabla_frecuencias = autos['body-style'].value_counts().reset_index()
tabla_frecuencias.columns = ["modelo", "Frecuencia"]
tabla_frecuencias = tabla_frecuencias.sort_values(by="modelo")

print("Tabla de Distribucion de Frecuencias")
tabla_frecuencias
```

```python
tabla_frecuencias["Frec Relativa"] = tabla_frecuencias["Frecuencia"] / 100
tabla_frecuencias
```

```python
tabla_frecuencias["Frec Porcentual"] = tabla_frecuencias["Frecuencia"] / 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias["Frec Acumulada"] = tabla_frecuencias["Frecuencia"].cumsum()
tabla_frecuencias
```
